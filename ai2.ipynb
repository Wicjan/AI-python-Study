```python
#p1n
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 0.2)
y = np.sin(x)

plt.grid(True)
plt.title('Wykres f(x) = sin(x)')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.plot(x,y,'b.')
plt.savefig('p1n.png', bbox_inches='tight', dpi=600)
```

```python
#p1s
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 0.2)
y = np.sin(x)

plt.axis([-1,0.8,-1,1])
plt.grid(True)
plt.title('Wykres f(x) = sin(x)')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.plot(x,y,'m.', markersize=10)
plt.savefig('p1s.png', bbox_inches='tight', dpi=600)
```

```python
#p2
import numpy as np
import matplotlib.pyplot as plt

a = 3
x = np.linspace(-10,10, 100)
y1 = a*np.sin(x-0.25)
y2 = a*np.sin(x-0.5)

plt.grid(True)
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.title("Dwa wykresy")
plt.axis([-2*np.pi,2*np.pi,-a-0.25,a+0.25])
plt.plot(x,y1, label='y = a * sin(t-0.25)')
plt.plot(x,y2,color='blue', label = 'y = a * sin(t-0.5)')
plt.legend()
plt.savefig('p2.png', bbox_inches='tight', dpi=600)
```

```python
#p4
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-10,10)
Y = np.arange(-10,10, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(X*X+Y*Y)

surf = ax.plot_surface(X, Y, Z, cmap=cm.seismic, linewidth=0,
antialiased=False)
ax.set_zlim(0, 15)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.savefig('p4.png', bbox_inches='tight', dpi=600)
```
